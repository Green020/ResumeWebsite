```python
import csv
from tabulate import tabulate
from prettytable import PrettyTable
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

datafilename = 'DiceRollData.csv'
dropdownfilename = 'DiceDataDropdowns.csv'

# Read the files
df = pd.read_csv(datafilename)
ddf = pd.read_csv(dropdownfilename)

# Display the data
# print(datafile)
# print(dropdownfile)

# Grab Data File Columns
df_ClassCol = df['Class']
df_CharNameCol = df['Character Name']
df_RollCol = df['Roll']
df_DieTypeCol = df['Die Type']
df_CheckSaveAttackCol = df['Roll Reason']
df_AttackRangeCol = df['Attack Range']
df_AttackTypeCol = df['Attack Type']
df_DamageDealtCol = df['Damage Dealt']
df_DamageTypeCol = df['Damage Type']
df_NaturalRollCol = df['Nat 1 or 20']

# Grab Dropdown Data File Columns
ddf_ClassCol = ddf['Class Dropdown']
ddf_DieTypeCol = ddf['Dice Type Dropdown']
ddf_CheckSaveAttackCol = ddf['Roll Reason Dropdown']
ddf_AttackRangeCol = ddf['Attack Range Dropdown']
ddf_AttackTypeCol = ddf['Attack Type Dropdown']
ddf_DamageTypeCol = ddf['Damage Type Dropdown']
ddf_NaturalRollCol = ddf['Nat 1 or 20 Dropdown']

# Read the CSV file
with open(datafilename, "r") as file:
    reader = csv.reader(file)
    headers = next(reader)  # Get the header row
    table = PrettyTable(headers)
    for row in reader:
        table.add_row(row)

# Display the table
print(table)

# Count The Values In Each Column
classCount = df_ClassCol.value_counts()["Monk"]

#Build The Bar Chart
plt.barh(ddf_ClassCol, classCount, color='violet')

plt.title('Class Pick')
plt.xlabel('Class')
plt.ylabel('Times Picked')
plt.show()
```
